```python
def Token(client_id,client_secret):
  token_url = "https://accounts.spotify.com/api/token"
  headers = {
      "Content-Type": "application/x-www-form-urlencoded",
  }
  data = {
      "grant_type": "client_credentials",
      "client_id": client_id,
      "client_secret": client_secret,
  }
  response = requests.post(token_url, headers=headers, data=data)
  if str(response)=='<Response [200]>':
    return response.json().get('access_token')
  else:
    return 0
```

```python
#Seconda versione

ARTISTS_ENDPOINT='https://api.spotify.com/v1/artists/'

def get_related_artists(token_var,id_1,artits,Artists):
  Artists_copy=Artists
  api_url = f"{ARTISTS_ENDPOINT}{id_1}/related-artists"
  headers = {
    "Authorization": f"Bearer {token_var}"}
  try:
    response = requests.get(api_url, headers=headers)
    print(response)
    if str(response)=='<Response [200]>':
      item=response.json().get('artists')
      t=0
      check=[]
      for i in item:
        id_2=str(i.get('id'))
        str_id_1=str(id_1)
        if id_2 in artists:
          t=t+1
          check.append(id_2)
          if str(id_1) not in Artists.loc[id_2,'RELATED_TO'].split(','):
            Artists_copy.at[id_2,'RELATED_TO']+=','+str(id_1)
          if str(id_2) not in Artists.loc[id_1,'RELATED_TO'].split(','):
            Artists_copy.at[id_1,'RELATED_TO']+=','+id_2
      h=1
      for i in check:
        if i in str(Artists_copy.at[id_1,'RELATED_TO']).split(','):
          h=h*1
        else:
          h=h*0

      Artists_copy.at[id_1,"RELATED_ARTIST_VER"]=t
      Artists_copy.at[id_1,"RELATED_ARTIST_ATT"]=str(len(str(Artists_copy.at[id_1,'RELATED_TO']).split(',')))+"-"+str(h)
      Artists_copy.at[id_1,"DONE"]=1
      Artists=Artists_copy
    elif str(response)=='<Response [429]>':
      return 429
    else:
      print(f"problema{response}")
      return 0
  except requests.exceptions.RequestException as e:
    print(f"Error in API request: {e}")
```
